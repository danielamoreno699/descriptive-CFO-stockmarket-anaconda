```python
sample_std = cfo_list.std()
sample_std
```

```python
sample_mean = cfo_list.mean()
sample_mean
```

```python
np.random.seed(123)
sample2 = np.random.normal(loc = sample_mean, scale = sample_std, size=10000)
sample2
```

```python
sample2_mean=sample2.mean()
```

```python
sample2_std= sample2.std()
```

```python
x = np.linspace(-20,30,10000)
```

```python
y = stats.norm.pdf(x,loc = sample2_mean, scale = sample2_std)
```

```python
plt.figure(figsize = (20,8))
plt.hist(sample2,bins=10000,density=True)
plt.plot(x,y,color = 'red', linewidth = 3)
plt.grid()
plt.title ('Normal Distribution', fontsize = 20)
plt.xlabel('cfo', fontsize = 15)
plt.ylabel('pdf', fontsize = 15)
plt.yticks(np.arange (0,1.1,0.11))
plt.show()
```
